```python
#Find the minima for (x+5)^2 using gradient descent
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# define the first value we are going to start from
cur_x = 3

# learning rate (alpha)
rate = 0.01

# precision ( difference between two consecutive values of x, used to stop the function to find gradient descent.)
precision = 0.00001

previous_step_size = 1
max_iterations = 10000

iters = 0

# diffrential of (x+5)^2
df = lambda x: 2*(x+5)

x_list = []
y_list = []

while previous_step_size > precision and iters < max_iterations:
    prev_x = cur_x
    cur_x = cur_x - rate * df(prev_x)
    previous_step_size = abs (cur_x - prev_x) # check if alpha is getting too small
    iters +=1 # increment the iterations until it reached 1000
#     print("Iteration:", iters, "Value of current minima is:", cur_x)
    x_list.append(cur_x)
    y_list.append(0)
    
    
# print("The value of the local minima is:", cur_x)

x_min = -20
x_max = 20

x = np.linspace(x_min, x_max)
y = np.linspace (-20, 20)

#plotting Gradient Descent
x_1 = np.linspace(-30, 20, 1000)
y_1 = (x_1+5)**2
plt.plot(x_1, y_1)
plt.scatter(x_list,y_list,color='#ff0023', label = 'Data Point')
plt.grid(True, linewidth='1', color='black')
plt.legend()
plt.show()





```
