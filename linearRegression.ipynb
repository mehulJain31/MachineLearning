```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('brain_weight.csv')

#relation between head size and brain weight
X = dataset['Head Size(cm^3)'].values
Y = dataset['Brain Weight(grams)'].values
n= len(X)
X_mean = np.mean(X)
Y_mean = np.mean(Y)

numerator = 0
denominator = 0

#get the values for Nr and Dr for least mean square
for i in range(n):
    numerator += (X[i] - X_mean)*(Y[i] - Y_mean)
    denominator += (X[i] - X_mean) **2

# get the LR equation m(scale/coeffecient) and b(bias coeffecient)    
#Brain weights = b0 + b1 * Head size
b1 = numerator/denominator
b0 = Y_mean - b1 * X_mean

#plotting the graph
x_max = np.max(X) + 100
x_min = np.min(X) - 100

# get the x axis with values between (x_max, x_min) spaced in increments of 1000
x = np.linspace(x_min, x_max, 1000)
y = b0 + b1 * x

#plotting line
plt.plot(x, y,color='#00ff00', label = 'Linear Regression')

# plot the datapoints
plt.scatter(X, Y, color='#ff0023', label = 'Data Point')

# X_axis label
plt.xlabel('Head Size')

#y_axis label
plt.ylabel('Brain Size')

plt.legend()
plt.show()

#calculate the accuracies
#1. Root Mean Squared Error
rmse = 0
for i in range(n):
    y_pred = b0 + b1 * X[i]
    rmse += (Y[i] - y_pred)**2
    
rmse = np.sqrt(rmse/n)
print("RMSE Accuracy is :", rmse)

# 2. Coefficient of Determination (R² Score)
sumofsquares = 0
sumofresiduals = 0
for i in range(n) :
    y_pred = b0 + b1 * X[i]
    sumofsquares += (Y[i] - Y_mean) ** 2
    sumofresiduals += (Y[i] - y_pred) **2
    
score  = 1 - (sumofresiduals/sumofsquares)

print("R² accuracy is: ", "%.2f" %(score*100), "%")
```
